```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("financial_risk_assessment.csv")
```

```
# Memilih kolom numerik dari data
numerical_data = df.select_dtypes(include=['float64', 'int64'])

# Menghitung statistik secara manual
mean = numerical_data.mean()
std_dev = numerical_data.std()
min_value = numerical_data.min()
max_value = numerical_data.max()
q1 = numerical_data.quantile(0.25)
q2 = numerical_data.median()  # Q2 sama dengan median
q3 = numerical_data.quantile(0.75)

# Menampilkan hasil
print("\nRata-rata (Mean):")
print(mean)
print("\nStandar Deviasi (Standard Deviation):")
print(std_dev)
print("\nNilai Minimum (Min):")
print(min_value)
print("\nNilai Maksimum (Max):")
print(max_value)
print("\nKuartil 1 (Q1):")
print(q1)
print("\nKuartil 2 / Median (Q2):")
print(q2)
print("\nKuartil 3 (Q3):")
print(q3)
```

```
# prompt: buatkan grafik visualisasi dataset age

# Membuat histogram untuk kolom 'age'
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()
```

```
# prompt: Tampilkan korelasi dataset menggunakan heatmap.

# Menghitung korelasi antar kolom numerik
correlation_matrix = numerical_data.corr()

# Membuat heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antar Variabel')
plt.show()
```

### Kesimpulan dari Hasil Analisis dan Visualisasi Data

1. Profil Demografi: Data ini mencakup individu dengan variasi usia, jenis kelamin, tingkat pendidikan, status pernikahan, dan status pekerjaan. Rata-rata usia bervariasi dari usia muda hingga tua, dengan berbagai tingkat pendidikan seperti Sarjana, Master, dan PhD.
2. Faktor Keuangan: Kolom seperti pendapatan, skor kredit, jumlah pinjaman, dan rasio utang-terhadap-pendapatan memberikan informasi penting tentang kondisi keuangan individu. Skor kredit bervariasi, yang mencerminkan risiko kredit yang berbeda di antara individu.
3. Status Pekerjaan dan Penghasilan: Status pekerjaan seperti 'Employed' dan 'Unemployed' memberikan gambaran stabilitas pendapatan. Mereka yang bekerja cenderung memiliki pendapatan yang lebih tinggi dan stabil dibandingkan dengan yang tidak bekerja.
4. Nilai Aset dan Jumlah Tanggungan: Nilai aset yang dimiliki individu bervariasi, dari aset yang kecil hingga besar. Jumlah tanggungan juga berpengaruh terhadap kapasitas finansial individu.
5. Riwayat Pembayaran dan Rasio Utang-Terhadap-Pendapatan: Riwayat pembayaran dan rasio utang-terhadap-pendapatan menjadi indikator penting dalam penilaian risiko. Individu dengan riwayat pembayaran yang buruk dan rasio utang tinggi lebih berisiko.
6. Penilaian Risiko: Berdasarkan data, individu dikelompokkan dalam berbagai kategori risiko seperti "Low", "Medium", dan "High". Faktor-faktor yang mempengaruhi risiko termasuk skor kredit, riwayat pembayaran, pendapatan, dan nilai aset.
7. Distribusi Geografis: Individu berasal dari berbagai negara, yang menunjukkan bahwa data ini bersifat internasional. Hal ini mungkin mempengaruhi analisis risiko karena perbedaan kebijakan kredit antar negara.
8. Perubahan Status Pernikahan: Variabel 'Marital Status Change' menunjukkan perubahan dalam status pernikahan, yang mungkin mempengaruhi kondisi keuangan individu secara signifikan.
